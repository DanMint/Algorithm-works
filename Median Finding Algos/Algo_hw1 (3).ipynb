##  Naive median finding algorithm 

The time Complexity of the Naive algorithm will be O(n * log(n)), however there is an exeption. If the array that we want to find the median on is sorted then the time complexity will be constant time. However in most cases the intial array wont be sorted thus the time complexity will be O(n * log(n)). The reason that it will be O(n * log(n)) is because to sort the array takes n*log(n), and it takes constant time to find the median.
Therefore, the time complexity is O(n * log(n)).

```python
import numpy as np
import matplotlib.pyplot as plt

x = [0, 10000000, 10000000*2,10000000*3,10000000*4,10000000*5,10000000*6,10000000*7,10000000*8,10000000*9,10000000*10]
y = [0, 649, 1102, 1703, 2265, 2846, 3537, 4054, 4554, 5124, 5740]

plt.plot(x, y)
plt.xlabel('Array length')
plt.ylabel('Time, ms')
```

## Toni Hoar Algorithm

The time complexity of Toni's Hoars algorithm is O(n). Meaning T(n) = O(n) + T(n/2).(we have O(n) because it takes linear time to calculate the rank and choose the pivot..). At the T(n/2) part Its important to note that it is not always the case n/2, if we get n/2 its a very good and lucky case. In the worst case it can be constant that does not depend on n, â€” the total time complexity will be O(n^2), lets proove it: our time complexity is dependent on how we devide our array when we choose the pivot. If we choose the pivot good every time then we will have linear time, However, if we will choose the pivot badley then we will have a time complexity of O(n^2). So for the worst time we will have: T(n) = n + (n - 1) + (n - 2) + ...... = aproximatly n^2 = O(n^2)
