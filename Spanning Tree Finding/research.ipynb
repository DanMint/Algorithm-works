# Prim's algorithm

Time complexity for Prims is havily dependent on the graph and the data structure that was used for the graph. In the worst time senario the time complexity will be O(E*logV). The reason for that is because we perform some linear operations in terms of E, and we perform O(E + V) operations with a binary heap (or a binary search tree): each time we dequeue the smallest edge from the priority queue, then, we have to traverse all the edges of the destination vertex ( O(E) ) and insert the edge into the priority queue ( O(logE) ). So, each iteration takes O(logE) = O(logV) time, so in total we have O((E + V) * log V) = O(E * logV) for connected graphs.

# Kruskal's algorithm

Assuming that we have E edges and V verticies the time complexity of Kruskals algorith will be O(E log E), or equivalently, O(E log V). Sorting the edges by weight using a comparison sort will take O(ElogE). We also need to make O(E) operations with DSU class. Time complexity for every DSU operation is O(logV), thus the total time complexity is O(ElogV).

# Boruvka's algorithm

In the each iteration of Boruvka's algorithm, we'll make O(E) operations with DSU and some constant-time operations, thus the complexity of every iteration is O(E*logV). We perform O(logV) iterations, because it can be shown that in every iteration, the amount of components decreases by two times, so it exponentially shrinks from V to 1. So, the total time complexity for my implementation of Boruvka's algorithm is O(E*log^2(V)).
