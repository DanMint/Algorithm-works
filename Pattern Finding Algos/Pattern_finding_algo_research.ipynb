##Naive implemintation

The time complexity of Naive Pattern Search method is O(m*n). The m is the size of pattern and n is the size of the main string.

```python
from matplotlib import pylab
from pylab import *

time = [10, 20, 30, 42, 50, 59, 69, 79, 89, 99, 193, 289, 188, 186, 181, 176, 178, 170, 167, 162]
string_size = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000, 20000, 20000, 20000, 20000, 20000, 20000, 20000, 20000, 20000, 20000]
pylab.plot(string_size, time)
pylab.title("Naive")
pylab.xlabel("String (length)")
pylab.ylabel("Time (ms)")

pylab.show()

```

##Rabin Karp


Assume the text is length n and the length of the word is m. The best- and average-case running time of Rabin-Karp is O(m + n)O(m+n) because the rolling hash step takes O(n)O(n) time and once the algorithm finds a potential match, it must verify each letter to make sure that the match is true and not a result of a hashing collision and therefore must check each of the m letters in the word.
The worst-case running time of Rabin-Karp is O(nm)O(nm). This would occur with an extremely awful hash function that resulted in a false positive at each step. Since whenever the algorithm thinks it found a match, it must verify each of the m letters in the word, if there is a collision at each step, m letters will be checked nn times resulting in a running time of O(nm)O(nm). This can be avoided with a good choice of hash function.


```python
from matplotlib import pylab
from pylab import *

time = [22, 38, 68, 73, 102, 109, 125, 143, 162, 180, 358, 391, 364, 353, 350, 348, 350, 349, 339, 347]
string_size = [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000, 20000, 20000, 20000, 20000, 20000, 20000, 20000, 20000, 20000, 20000]
pylab.plot(string_size, time)
pylab.title("Rabin karp")
pylab.xlabel("String (length)")
pylab.ylabel("Time (ms)")

pylab.show()
```
